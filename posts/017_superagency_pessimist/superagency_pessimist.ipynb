I started Superagency with my guard up. Not in a dramatic way, but in the quiet, habitual way you develop after spending enough time reading about AI. I have read enough papers, headlines, and policy memos to instinctively flinch when someone tells me technology will “empower” us. Empowerment has a habit of meaning efficiency for institutions and precarity for individuals. So when Superagency framed AI as a tool that could expand human agency, I read it with suspicion rather than hope.

What I appreciated immediately, though, was that the book does not pretend the risks are imaginary. It acknowledges that AI systems can centralize power, entrench inequality, and obscure accountability. That said, my pessimism kept interrupting the narrative. Every optimistic claim triggered a counterfactual in my head. Who controls these systems. Who benefits first. Who is left adapting rather than shaping. I kept thinking about how often agency is framed as something individuals should exercise, even when the structural conditions make meaningful choice impossible. It seemed like every piece of good described in the book completely oversimplified all the bad.

The book argues that agency is not something AI automatically strips from us. Instead, it depends on how systems are designed, governed, and distributed. I intellectually agree with this, but emotionally, I struggled. The history of technology does not exactly reward trust. We rarely design systems with the most vulnerable users in mind. We usually retrofit safeguards after harm becomes visible and unavoidable. Those who benefit from AI the most monetarily cannot be trusted to actually build these safeguards. Times Person of the Year was basically the most wealthy leaders in the world, all working on AI. If the book believes that the goodness of AI is left to them, I am extremely skeptical.

Still, Superagency forced me to articulate why I am pessimistic rather than letting that stance remain a reflex. I realized my skepticism is less about AI itself and more about institutions. I do not doubt that AI could augment human decision making, creativity, and coordination. I doubt our willingness to prioritize those outcomes over profit, speed, and control.

By the end of the book, I was not convinced that AI will make us more agentic by default. Pessimism can be useful when it sharpens critique, but it becomes dangerous when it excuses disengagement. I closed the book still wary, still unconvinced by many of its assumptions, but slightly more uncomfortable with my own certainty that things will inevitably go wrong. I still think it was productive to read the book with a pessimistic mindset at times (and an optimist at others, coming soon in another blog), but I am not sure that I was converted fully!
