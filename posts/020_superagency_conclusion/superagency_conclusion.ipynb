My reasoning for choosing to read SUperagency may have been different than others. When we were given the choice to rank books based on how much we wanted to read them, and were given this wide scale betweeen pro and anti AI books, it took me some time to decide what I wanted to do. I have always been a gloomer and doomer about AI and what it is doing to the human mind and psyche, so the clear choice was to read books by authors who were also worried about this as well as bias in AI systems. However, I was presented a chance to flip the narrative for myself. While I was wary about this, I was excited about the opportunity to force myself way out of my comfort zone. At the very least, I would be able to understand the other side.

What makes this even more complex for me is that I am a Computer Science major. This means that on a daily basis, I do not just battle what AI does to society but I battle what it does for my education and for my job. So while I remain negative on the subject, I thought it would be nice to shine some positivity on the subject. At the end of the semester, regardless of how much I enjoyed reading the book, it did do that for me. The book made it harder for me to hold onto simplistic narratives about AI, whether optimistic or pessimistic.

Before reading it, I tended to frame AI in terms of risk management. What could go wrong. Who might be harmed. How systems fail. Those questions still matter deeply to me, but the book pushed me to also ask what could go right, and under what conditions.

The bookâ€™s arguments depend heavily on the assumption that we will choose to design AI in ways that expand agency broadly rather than concentrate it. History gives me reasons to doubt that assumption. However, if anything, the book made me more interested in reading further. Not just about AI capabilities, but about governance, institutional design, and historical analogies. I want to read books that interrogate who gets to define agency in the first place. And read more from authors with this perspective. The authors of this book are so deep rooted in AI (in a financial sense as well), so it would be nice to read a pro-AI book by authors that do not depend on it for their livelihood. Because, at the end of the day, this book was a marketing tactic for their business lives!

I am excited, but with conditions. Excited by the possibility that AI could help people think better, learn faster, and act more deliberately. I left with better questions, and a clearer sense that the future of AI is not something that happens to us. It is something we participate in, whether we intend to or not.
