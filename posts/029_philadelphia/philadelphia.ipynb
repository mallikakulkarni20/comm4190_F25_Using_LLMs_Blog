# Background

Philadelphia is a city that resists simple summaries. It is historic without feeling frozen in time, gritty without being unlivable, and deeply academic without being sealed off from the rest of the world. It is a place where museums exist next to food trucks, where college campuses bleed directly into city neighborhoods, and where local culture is loud, opinionated, and extremely specific.

As a college student, my experience of Philadelphia is shaped by walking everywhere, discovering places through friends rather than lists, and slowly learning which parts of the city feel like mine. Philly is not a place you “do” in a weekend. It is a place you grow into.

Because of this, I was curious how an LLM would describe Philadelphia. Unlike Johns Creek, Philly generates enormous amounts of online content. It has history, tourism, nightlife, universities, sports, and strong opinions. On paper, this should make it easier for an LLM to recommend activities accurately.

But accuracy is not the same as understanding. I wanted to see whether an LLM could capture the version of Philadelphia that college students actually experience, or whether it would default to a tourist-facing version of the city.

---

# Setup

I used Gemini 2.5 Flash and asked it a series of prompts about activities in Philadelphia. I did not specify that I was a student at first. I wanted to see the default representation of the city.

Then, I adjusted the prompt to include student context and compared the results.

Each prompt was run independently.

---

## Prompt 1

**“What are some things to do in Philadelphia, Pennsylvania?”**

### Result

Gemini immediately listed major landmarks and attractions. It mentioned historical sites, museums, famous food spots, and sports culture. The recommendations were broad and recognizable.

This response was accurate but generic. It felt like a travel brochure. It reflected what Philadelphia is known for rather than how it is lived in.

There was no sense of scale, routine, or daily life.

---

## Prompt 2

**“What cultural activities are popular in Philadelphia?”**

### Result

Gemini responded with museums, theaters, music venues, and festivals. It referenced art scenes, historical preservation, and diverse neighborhoods.

This answer was more textured than the first, but it still framed culture as something people attend rather than something people participate in. Culture was presented as events and institutions, not habits or communities.

Philly’s culture often shows up in small moments. Late-night food runs. Neighborhood loyalty. Debates about which places are overrated. None of that appeared.

---

## Prompt 3

**“What are fun things to do in Philadelphia for college students?”**

### Result

This answer shifted noticeably. Gemini mentioned cafes, study-friendly spots, nightlife, budget-friendly food, parks, and student-heavy neighborhoods. It referenced walking, exploring, and social spaces.

This was the closest the model came to capturing my experience of the city. The tone became more practical and relational. It acknowledged that students interact with the city differently than tourists.

However, the recommendations still felt generalized. They were correct, but they lacked specificity. It was the idea of a student city, not *this* student city.

---

# Analysis

Gemini performs much better with Philadelphia than with Johns Creek, but the difference is revealing.

## What Gemini Does Well

- **Visibility-based accuracy**  
  Philly is well-documented online. Gemini has access to abundant descriptions of what exists.

- **Tourist and institutional framing**  
  It accurately describes what the city is known for publicly.

- **Context switching**  
  When prompted with “college students,” it adjusts tone and recommendations appropriately.

## What Gemini Still Misses

- **Scale of lived experience**  
  College life in Philly revolves around walking routes, repeat spots, and familiar neighborhoods. This repetition is missing.

- **Social discovery**  
  Students often find places through friends, not search engines. Gemini does not capture this organic discovery process.

- **Neighborhood identity**  
  Philly is intensely neighborhood-based. Gemini treats the city as a unified space rather than a patchwork of micro-communities.

- **Emotional attachment**  
  The model does not capture how places become meaningful over time. A random coffee shop becomes important because of who you went with, not because it is highly rated.

---

# Comparing Philly to Johns Creek

What stood out most was the contrast between how Gemini treated my hometown versus my college city.

Johns Creek was evaluated through metrics.  
Philly was evaluated through landmarks.

In both cases, the model relied on externally visible data. What changes is the type of data available. Suburbs are flattened into safety and livability. Cities are flattened into attractions and culture.

Neither approach fully captures lived experience.

---

# Why This Matters

As LLMs are increasingly used for recommendation, exploration, and planning, this gap matters. People may rely on AI to decide where to go, what to do, and how to experience a place.

But places are not lists. They are routines, relationships, and repetition.

Philadelphia, as I know it, is shaped by:

- walking to class  
- eating the same cheap food weekly  
- discovering new streets slowly  
- building emotional landmarks rather than tourist ones  

Gemini can recommend Philly. It cannot replicate growing into it.

---

# Takeaways

Gemini does a good job describing Philadelphia as a destination. It does a decent job adjusting for college students when prompted. But it still frames the city through visible, searchable experiences rather than personal ones.

This highlights a broader truth about LLMs. They are excellent at summarizing public narratives. They are weaker at capturing private routines.

Philadelphia is not just a city you visit. It is a city you slowly learn. That learning process, the quiet accumulation of familiarity, is something LLMs do not experience and therefore struggle to describe.
