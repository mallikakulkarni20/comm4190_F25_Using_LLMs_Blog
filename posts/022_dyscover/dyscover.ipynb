# Background

I wanted to share a bit about a hackathon I completed last semester, where the prompt was to build an AI tool for EdTech. THis hackathon was particularly important to me because my team and I not only won first place, but also got the opportunity to talk to may influential people at Wharton's AI and Analytics Initiative Team and GSE at Penn!

# About Dyscover
My biggest accomplishment to date has been winning first place at the 2025 Wharton Hack-AI-thon with my team’s project, Dyscover. The challenge asked us to create a tool that could revolutionize education, and we built an AI handwriting analysis model designed for early dyslexia detection. Rather than designing another AI tutor, we focused on empowering teachers — giving them both a diagnostic tool and a GPT-4 powered lesson plan generator tailored to each student’s needs.


The hackathon itself was intense, with multiple all-nighters and a 9 a.m. final pitch after working until sunrise. Presenting to the Wharton AI and Analytics Initiative Advisory Board (an audience of 20+ alumni, CEOs, and founders) was an incredible experience to have as a young student. Their response opened doors for us well beyond the competition.

The tech itself was created by using Google's Teachable Machine to train a CNN on open-source dyslexic/non-dyslexic handwriting samples, exported as TensorFlow. Why this tech stack? Teachable Machine allowed rapid prototyping (crucial for a hackathon), CNNs excel at pattern recognition for subtle visual features that are hard to define explicitly, and TensorFlow ensured scalability for future deployment. We integrated OpenAI's API for generating personalized lesson plans, with predefined prompts for ease of use. Also, we used Gradio UI for intuitive teacher interaction—upload sample, receive confidence score.

In the weeks that followed, I was able to connect with leaders in the learning disability space, from the Chief of Data and AI at the United Nations International Computing Center to educational researchers at AIM Academy. We took the initiative to work with researchers at New York colleges for data collection and learned a lot from experts in the field. We often interact with the Wharton AI and Analytics Initiative and received $5k+ to attend the Grace Hopper Celebration.


This accomplishment means more to me than just winning a hackathon. It represents the ability to transform an idea into something tangible, to communicate it effectively to experts, and to spark collaborations that could extend its reach. For me, it’s proof that I can combine technical skills, teamwork, and vision to build something that matters.
