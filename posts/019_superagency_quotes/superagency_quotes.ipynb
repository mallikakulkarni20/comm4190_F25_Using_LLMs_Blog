Some books change how you think. Others change the sentences that replay in your head. Superagency did both, but what surprised me most was how many lines I underlined. Not because they were dramatic, but because they were precise. They put language to intuitions I had not fully articulated.

One recurring theme I kept highlighting was the idea that agency is structured. The book repeatedly emphasizes that our capacity to act meaningfully depends on the systems around us. That framing felt especially relevant in conversations about AI, where responsibility is often misplaced onto individual users rather than designers or institutions.

I was drawn to passages that rejected technological determinism. Lines that stressed that AI does not arrive with fixed social consequences, but acquires them through deployment choices. That idea is not new, but it is stated here with unusual clarity. It cuts through both hype and panic by insisting that outcomes are contingent.

Another set of highlights focused on augmentation. I loved the emphasis on AI as a collaborator rather than a competitor. The language around “expanding the action space” stuck with me. It reframes intelligence not as a zero-sum resource, but as something that can be distributed and scaffolded.

There were also quieter moments that stayed with me. Reflections on how tools subtly reshape what we consider possible. How convenience can narrow agency if it removes opportunities for deliberation. Those sections felt less like a manifesto and more like a warning written in lowercase.

I appreciated that the book does not shy away from tradeoffs. Some of my favorite quotes acknowledged that increasing agency for some can reduce it for others if systems are poorly designed. That honesty made the optimistic claims more credible.

What I enjoyed most overall was the tone. It was confident without being dismissive, hopeful without being saccharine. The quotes I marked were not trying to persuade through spectacle. They were trying to clarify.

By the end, my margins were crowded. Not with rebuttals, but with connections. To things I have studied, built, worried about. That is usually how I know a book worked. Not because it convinced me of everything, but because it gave me better language to argue with myself.
